```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
img= cv2.imread("/content/drive/MyDrive/Computer_vision/Traditional Computer Vision/Image Enhancement/lena.jpg")

cv2_imshow(img)
```

```python
plt.hist(img.ravel(), 256, [0, 256])
```

```python
b, g, r= cv2.split(img)

cv2_imshow(b)
cv2_imshow(g)
cv2_imshow(r)
```

```python
plt.hist(b.ravel(), 256, [0, 256])
plt.hist(g.ravel(), 256, [0, 256])
plt.hist(r.ravel(), 256, [0, 256])
```

```python
hist= cv2.calcHist([img], [0], None, [256], [0, 256])      # [0]: means first channel

plt.plot(hist)

plt.show()
```

```python
colors= ['b', 'g', 'r']

for i,color in enumerate(colors):
    hist = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(hist,color = color)
plt.title('Image Histogram GFG')
plt.show()
```
