```python
import cv2
import matplotlib.pyplot as plt

from google.colab.patches import cv2_imshow
```

```python
img= cv2.imread("/content/drive/MyDrive/Computer_vision/Traditional Computer Vision/Corner Detection/download.jpg", 0)

params= cv2.SimpleBlobDetector_Params()           # el function di eli b7dd biha el parameters

params.minThreshold = 0
params.maxThreshold = 255

params.filterByArea = True
params.minArea = 500
params.maxArea = 2000

params.filterByColor = False
params.blobColor = 0

# filter by radius
params.filterByCircularity = True
params.minCircularity = 0.1
params.maxCircularity = 1

# convix shape
params.filterByConvexity = True
params.minConvexity = 0.1
params.maxConvexity = 1
```

```python
detector= cv2.SimpleBlobDetector_create(params)                # function to detect blob

keypoints= detector.detect(img)

print(f"Number of blob detected are: {len(keypoints)}")

img_with_blobs= cv2.drawKeypoints(img, keypoints, None, (0, 0, 255), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

cv2_imshow(img_with_blobs)
```
