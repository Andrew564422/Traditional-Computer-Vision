```python
import cv2
import mediapipe as mp

from google.colab.patches import cv2_imshow
```

```python
img= cv2.imread("/content/drive/MyDrive/Computer_vision/Traditional Computer Vision/Object detection/face detection using haar cascade/Andrew.jpg")

cv2_imshow(img)
```

```python
face_mesh= mp.solutions.face_mesh.FaceMesh()  # facemesh msjh btsht8l 8er lw el sora rgb msh gbr

rgb_img= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

result= face_mesh.process(rgb_img)

height, width, _ = img.shape

for facial_landmarks in result.multi_face_landmarks:         # if i have more than one face
  for i in range(len(facial_landmarks.landmark)):
    pt= facial_landmarks.landmark[i]                         # Normalized point

    print(pt)

    x= int(pt.x * width)
    y= int(pt.y * height)

    cv2.circle(img, (x, y), 2, (100, 100, 0), -1)

cv2_imshow(img)
```
