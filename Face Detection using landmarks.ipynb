```python
import dlib
import numpy as np
import cv2

from google.colab.patches import cv2_imshow
```

```python
detector= dlib.get_frontal_face_detector()                   # Face detection
predictor= dlib.shape_predictor("/content/drive/MyDrive/Computer_vision/Traditional Computer Vision/Object detection/shape_predictor_68_face_landmarks.dat")                           # landmarks drawn
```

```python
img= cv2.imread("/content/drive/MyDrive/Computer_vision/Traditional Computer Vision/Object detection/Andrew.jpg")
grey= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2_imshow(grey)
```

```python
faces= detector(grey)          # this function returns 4 values for top left and bottom right

for face in faces:
  x1= face.left()
  y1= face.top()

  x2= face.right()
  y2= face.bottom()

  cv2.rectangle(img, (x1,y1), (x2,y2), (0, 255, 0), 3)

  landmarks= predictor(grey, face)    # hna bdilo el image kolha w b7ddlo mkan el face bzabt eli gbto abl kda
  for n in range(0, 68):
    x= landmarks.part(n).x
    y= landmarks.part(n).y

    cv2.circle(img, (x,y), 3, (0, 255, 0), -1)
```
