```python
import numpy as np
import cv2

from google.colab.patches import cv2_imshow
```

```python
messi= cv2.imread("/content/drive/MyDrive/Computer_vision/Traditional Computer Vision/Object detection/Messi.jpg")
grey_img= cv2.cvtColor(messi, cv2.COLOR_BGR2GRAY)

cv2_imshow(grey_img)
```

```python
grey_img.shape
```

```python
messi_face= grey_img[0:70, 150:200]

cv2_imshow(messi_face)
```

```python
w, h= messi_face.shape[::-1]    #-1 reverse returns of shape function from h, w, c to w, h, c
```

```python
res= cv2.matchTemplate(grey_img, messi_face, cv2.TM_CCOEFF_NORMED)

cv2_imshow(res)

print(res)
```

```python
thr= 0.99
loc= np.where(res >= thr)

print(loc)
```

```python
for pt in zip(*loc[::-1]):
  cv2.rectangle(messi, pt, (pt[0] + w, pt[1] + h), (0, 0, 255), 2)

cv2_imshow(messi)
```
