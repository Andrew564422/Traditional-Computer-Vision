```python
import cv2
import time
```

```python
#line a
ax1= 70
ay= 90
ax2= 230

#line b
bx1= 15
by= 125
bx2= 225
```

```python
def calc_speed(time):
  try:
    speed= (9.144/1000) / (time/3600)
    return speed
  except ZeroDivisionError:
    print("can not devide by zero")
```

```python
i = 0         # Car count

start_time= time.time()       # calculate start time

detector= cv2.CascadeClassifier("/content/drive/MyDrive/Computer_vision/Traditional Computer Vision/Object detection/calculate car speed/cars1.xml")

cap= cv2.VideoCapture("/content/drive/MyDrive/Computer_vision/Traditional Computer Vision/Object detection/calculate car speed/video3.mp4")

fourcc= cv2.VideoWriter_fourcc('X', 'V', 'I', 'D')  # for quality, resolution, and colours
out= cv2.VideoWriter("/content/drive/MyDrive/Computer_vision/Traditional Computer Vision/Object detection/calculate car speed/out_video.avi", fourcc, 30, (1280, 720))
```

```python
while True:
  ret, img= cap.read()

  if not ret:
    break

  blur= cv2.blur(img, (15, 15))
  grey= cv2.cvtColor(blur, cv2.COLOR_BGR2GRAY)
  cars= detector.detectMultiScale(grey, 1.1, 2)

  cv2.line(img,(ax1,ay),(ax2,ay),(255,0,0),2)
  cv2.line(img,(bx1,by),(bx2,by),(255,0,0),2)

  for (x, y, w, h) in cars:
    cv2.rectangle(img, (x,y), (x+w, y+h), (0, 0, 255), 2)
    cv2.circle(img, (int((x + x + w)/2), int((y + y + h)/2)), 1, (0, 255, 0), -1)

    while int(ay) == int((y + y + h)/ 2):
      cv2.line(img,(ax1,ay),(ax2,ay),(0,255,0),2)
      start_time= time.time()
      break

    while int(ay) <= int((y + y + h)/ 2):
      if int(by) <= int((y+y+h)/2) & int( by + 10 ) >= int(( y + y+h)/2):
        cv2.line(img,(bx1,by),(bx2,by),(0,255,0),2)
        Speed = calc_speed(time.time() - start_time)
        print("Car Number "+str(i)+" Speed: "+str(Speed)+ " KM/H ")
        i = i + 1
        break
      else:
        cv2.putText(img, "Calcuting", (100,200), cv2.FONT_HERSHEY_SIMPLEX,1,(0,255,0),3)
        break

  image= cv2.resize(img, (1280, 720))
  out.write(image)

  if cv2.waitKey(33) == 27:
      break

cap.release()                  #close cam
out.release()
cv2.destroyAllWindows()        #close window
```
