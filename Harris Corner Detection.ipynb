```python
import cv2
import numpy as np

from google.colab.patches import cv2_imshow
```

```python
img= cv2.imread("/content/drive/MyDrive/Computer_vision/Traditional Computer Vision/Harris Corner Detection/Harris-Corner.jpg")

cv2_imshow(img)

grey= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

grey= np.float32(grey)                # Harrison must work on float32 img

dst= cv2.cornerHarris(grey, 2, 3, 0.04)      # (img, stride of box, kernel size, k in rule)

cv2_imshow(dst)       # strong and weak corners

img[dst > 0.01* dst.max()] = [0, 0, 255]

cv2_imshow(img)       # strong corners only
```

```python
img= cv2.imread("/content/drive/MyDrive/Computer_vision/Traditional Computer Vision/Harris Corner Detection/Harris-Corner.jpg")

cv2_imshow(img)

grey= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

grey= np.float32(grey)                # Harrison must work on float32 img

dst= cv2.cornerHarris(grey, 2, 3, 0.04)      # (img, stride of box, kernel size, k in rule)

dst = cv2.dilate(dst, None)        # to increas thickness of pixel to make corners more clear

cv2_imshow(dst)       # strong and weak corners

img[dst > 0.01* dst.max()] = [0, 0, 255]

cv2_imshow(img)       # strong corners only
```
