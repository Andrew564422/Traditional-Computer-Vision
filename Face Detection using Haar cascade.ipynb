## Haar Cascade Classifier

### https://towardsdatascience.com/face-detection-with-haar-cascade-727f68dafd08

### it is a machine leraning based approach

### this algorithm has 4 main steps
- Selecting Haar-like features: sliding fatures on image
- Creating an integral image:  It is used as a quick and efficient way to calculate the sum of pixel values in an image or rectangular part of an image.
- Running AdaBoost training
- Creating classifier cascades

```python
import numpy as np
import cv2

from google.colab.patches import cv2_imshow
```

```python
img= cv2.imread("/content/drive/MyDrive/Computer_vision/Traditional Computer Vision/Object detection/Andrew.jpg")
grey_img= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2_imshow(img)
cv2_imshow(grey_img)
```

```python
face_cascade =cv2.CascadeClassifier("/content/drive/MyDrive/Computer_vision/Traditional Computer Vision/Object detection/haarcascade_frontalface_default.xml")
eyes_cascade =cv2.CascadeClassifier("/content/drive/MyDrive/Computer_vision/Traditional Computer Vision/Object detection/haarcascade_eye.xml")

faces= face_cascade.detectMultiScale(grey_img, 1.5, 4)   # (img, reduce factor for image size, # of neighbor boxes for adagrad algorithm)
eyes= eyes_cascade.detectMultiScale(grey_img, 2, 4)

for (x, y, w, h) in faces:
  cv2.rectangle(img, (x, y), (x + w, y + h), (0, 0, 255), 2)
  for (x, y, w, h) in eyes:
    cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)

cv2_imshow(img)
```
