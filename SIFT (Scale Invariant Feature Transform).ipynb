```python
import numpy as np
import cv2 as cv
from google.colab.patches import cv2_imshow

# Detection only
img = cv.imread('/content/drive/MyDrive/Computer_vision/Traditional Computer Vision/Corner Detection/OIP.jpg')
gray= cv.cvtColor(img,cv.COLOR_BGR2GRAY)

sift = cv.SIFT_create()
kp = sift.detect(gray,None)

img=cv.drawKeypoints(gray,kp,img)

cv2_imshow(img)

# cv.imwrite('sift_keypoints.jpg',img)
```

```python
import numpy as np
import cv2 as cv
from google.colab.patches import cv2_imshow

# Detection only
img = cv.imread('/content/drive/MyDrive/Computer_vision/Traditional Computer Vision/Corner Detection/OIP.jpg')
gray= cv.cvtColor(img,cv.COLOR_BGR2GRAY)

sift = cv.SIFT_create()
kp = sift.detect(gray,None)

img=cv.drawKeypoints(gray,kp,img, flags= cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)     # di btrsmli el daira bna2n 3la el size w el orientation w el scale

cv2_imshow(img)

# cv.imwrite('sift_keypoints.jpg',img)
```

```python
# Detection and description
img = cv.imread('/content/drive/MyDrive/Computer_vision/Traditional Computer Vision/Corner Detection/OIP.jpg')
gray= cv.cvtColor(img,cv.COLOR_BGR2GRAY)

sift = cv.SIFT_create()
kp, des = sift.detectAndCompute(gray,None)

img=cv.drawKeypoints(gray,kp,img)

cv2_imshow(img)
```
