## Fast Corners

### Fast Corners tries to get interest points (which have maximum information)

### it is faster than harris corners, bec it based onmachine learning (decision tree)

```python
import cv2
import numpy as np

from google.colab.patches import cv2_imshow
```

```python
# with nonmaxsupression

img= cv2.imread("/content/drive/MyDrive/Computer_vision/Traditional Computer Vision/Corner Detection/R.jpg")

fast= cv2.FastFeatureDetector_create()

kp= fast.detect(img, None)

image= cv2.drawKeypoints(img, kp, None, color=(255, 0, 0))

cv2_imshow(image)

print(f"total number of points with nonmaxsupression: {len(kp)}")
```

```python
# without nonmaxsupression

fast.setNonmaxSuppression(0)

kp= fast.detect(img, None)

image= cv2.drawKeypoints(img, kp, None, color=(255, 0, 0))

cv2_imshow(image)

print(f"total number of points with nonmaxsupression: {len(kp)}")
```
